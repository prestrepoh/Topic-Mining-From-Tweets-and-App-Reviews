```python
import pandas as pd
```

```python
data_location = root_PATH + '/data/dataset_aire_2019_stanik_et-al/dataset.csv'
data_export_location = root_PATH + '/data/preprocessed-AIRE-dataset'
```

```python
data_df = pd.read_csv(data_location)
```

## Replace all mentions (@MyCorp, etc) with the token: "@mention"

(This token will be added to the tokenizer)

```python
#Replace all mentions with @mention
data_df['text'] = data_df['text'].replace(to_replace='\@(\w+)',value='@mention',regex=True)
```

## Create the target column

```python
#Create target column
def extract_target(row):
    
    category = row['category']
    
    result = 0
    if category == 'irr':
        result = 0
    elif category == 'pbr':
        result = 1
    elif category == 'inq':
        result = 2
    return result


data_df['target'] = data_df.apply(lambda row: extract_target(row), axis=1)
```

## Export Preprocessed dataset

```python
data_df.to_csv(data_export_location + '/dataset-preprocessed.csv', index=False)
```
